```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from pathlib import Path
```

Data pulled from: https://www.cancer-rates.info/tx/

Years 2015-2019 * Breast Cancer (Either Sex) * County Level


```python
#Years 2015-2019 * Breast Cancer (Either Sex) * County Level * ALL Race/Ethnicity

csv_path = Path("CSV_Files/bcancertexas.csv")
breast_cancer_texas = pd.read_csv(csv_path, skiprows=2)
bct_clean= breast_cancer_texas.replace('~',0, regex=True)
bct_clean_drop=bct_clean.drop([246,247, 248, 249])
bct_clean_drop["Cases"]= bct_clean_drop["Cases"].astype(int)
bct_clean_drop["Cases"]
y_var= bct_clean_drop.loc[bct_clean_drop["Cases"]>1000]
y_var

#Bar graph of counties in texas that have over 1000 cases of breast cancer 2015-2019
plt.bar(y_var["County"], y_var["Cases"], color="pink")
plt.xticks(rotation=90)
plt.title("Counties in Texas with over 1,000 cases of Breast Cancer")
```

```python
cases_df= pd.DataFrame({"Race" : ["Native American & Alaskan", 
                                  "White",
                                  "Asian & Pacific Islander", 
                                  "Black", 
                                  "Hispanic"], 
                        "Total" : [bct_NA_clean["Cases"].sum(), 
                                   bct_NHW_clean["Cases"].sum(),
                                   bct_API_clean["Cases"].sum(),
                                   bct_NHB_clean["Cases"].sum(),
                                   bct_H_clean["Cases"].sum(), 
                                   ]})

cases_df
```

```python
fig1, ax1 = plt.subplots()
plt.pie(cases_df["Total"], labels=cases_df["Race"], autopct="%1.1f%%", startangle=0)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.title("Breast Cancer Incidences by Race 2015-2019 (Combined)")
plt.show()
```

```python
combined_year_list = []

def combine_race (df, race):
    
    df["Race"] = race
    combined_year_list.append(df)

combine_race(bct_clean_drop,"All Races")
combine_race(bct_NHW_clean, "Non-Hispanic White")
combine_race(bct_NHB_clean, "Non-Hispanic Black")
combine_race(bct_H_clean,  "Hispanic")
combine_race(bct_API_clean, "Asian & Pacific Islander")
combine_race(bct_NA_clean, "Native American & Alaskan")

combined_race_df= pd.concat(combined_year_list)
combined_race_df
```
